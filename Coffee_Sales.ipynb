```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
#
```

```python
df = pd.read_csv('/content/index.csv')
df.info()
```

```python
df.describe()
```

```python

df.head()
```

```python
plt.figure(figsize=(18,12))
sns.barplot(x = df['coffee_name'],y = df['money'],data=df,palette='magma')

plt.show()
```

```python
sns.barplot(x = df['cash_type'],y = df['money'].count(),data=df,palette='viridis')
plt.ylabel('Total Money')
plt.xlabel('Cash Type')
plt.show()
```

```python
plt.figure(figsize=(18,12))
m = df['date']
n = df['money']
plt.hist(m,color='green',bins=122,edgecolor='black')
plt.xlabel('Date [01/03/2024-30/06/2024]')
plt.ylabel('Count(Orders)')
plt.title('Day-wise Customer Count Analysis')
plt.show()
```
